```python
data = pd.read_csv("data/processed/chatgpt_data.csv")
data.head()
```

### 0. Gender

```python
temp = pd.get_dummies(data["Gender"])
data = pd.concat([data, temp], axis = 1).reindex(data.index)
data.drop("Gender", axis = 1, inplace = True)
data.head()
```

### 1. Nationality

```python
data["Nationality"].value_counts()
```

```python
temp = pd.get_dummies(data["Nationality"])
data = pd.concat([data, temp], axis = 1).reindex(data.index)
data.drop("Nationality", axis = 1, inplace = True)
data.head()
```

### 2. Genre/Niche

```python
temp = pd.get_dummies(data["Genre_Niche"])
data = pd.concat([data, temp], axis = 1).reindex(data.index)
data.drop(["Genre_Niche", "Genre_Niche_1","Genre_Niche_2"], axis = 1, inplace = True)
data.head()
```

### 3. Skills

```python
temp = pd.get_dummies(data["Skill_1"])
data = pd.concat([data, temp], axis = 1).reindex(data.index)
temp = pd.get_dummies(data["Skill_2"])
data = pd.concat([data, temp], axis = 1).reindex(data.index)
temp = pd.get_dummies(data["Skill_3"])
data = pd.concat([data, temp], axis = 1).reindex(data.index)
data.drop(["Skill_1", "Skill_2", "Skill_3", "Skills"], axis = 1, inplace = True)
data.head()
```

```python
data = data.loc[:,~data.columns.duplicated()].copy()
data.head()
```

### 4. Education

```python
data["Education"].value_counts()
```

```python
# self determined general degrees
education_options = ["Self-Taught", "High School Diploma", "Culinary School", "Bachelor's", "Master's"]
```

```python
def change_education(x):
    edu = x["Education"]
    op = None
    for i in education_options:
        if i in edu:
            op = i
    return op
    

data["General_Education"] = data.apply(change_education, axis = 1)
data.head()
```

```python
temp = pd.get_dummies(data["General_Education"])
data = pd.concat([data, temp], axis = 1).reindex(data.index)
data.drop(["Education", "General_Education"], axis = 1, inplace = True)
data.head()
```
