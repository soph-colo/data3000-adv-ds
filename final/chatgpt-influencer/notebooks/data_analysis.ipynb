# Data Analysis

#### Summary: This notebook will isolate one variable at a time and show the relationship between that variable and the income per year. This is just to take a look at what ChatGPT understood to be indicators of a money maker. Also included is a snapshot of the coefficents dataset so you can see what my model saw as important.

### Imports

```python
import pandas as pd
import os
from matplotlib import pyplot as plt
```

```python
p = os.chdir("..")
coeff = pd.read_csv("data/processed/coefficents.csv")
data = pd.read_csv("data/processed/chatgpt_data.csv")
condensed = pd.read_csv("data/processed/chatgpt_data_final.csv")
```

```python
coeff = coeff.sort_values(by = ['1'])
coeff
```

#### 0. Nationality

```python
nationality_income = data.groupby(['Nationality'])['Income_Per_Year'].mean().to_frame().reset_index()
nationality_income = nationality_income.sort_values(by =['Income_Per_Year'])
X = nationality_income['Nationality']
y = nationality_income['Income_Per_Year']
plt.figure(figsize=(20,10))
plt.xticks(rotation=90)
plt.title("Average Income Based On Nationality")
plt.bar(X, y)
```
